```python
import unidecode
import contractions
import re
from nltk.tokenize import word_tokenize
#Spelling correction
from textblob import Word
from nltk.corpus import stopwords

import nltk
nltk.download('punkt')

nltk.download('stopwords')
```

```python
df['PREPROCESSED_TEXT'] = df['TEXT']
df
```

```python
#removing @mentions
df['PREPROCESSED_TEXT'] = df['PREPROCESSED_TEXT'].apply(lambda x: re.sub(r'@[\w]*', '', x))
df
```

```python
#Removing URL's
df['PREPROCESSED_TEXT'] = df['PREPROCESSED_TEXT'].apply(lambda x: re.sub(r'https?:\/\/\S+', '', x))
df['PREPROCESSED_TEXT'] = df['PREPROCESSED_TEXT'].apply(lambda x: re.sub(r"www\.[a-z]?\.?(com)+|[a-z]+\.(com)", '', x))
df
```

```python
def remove_whitespace(text):
    # remove extra whitespaces from text
    text = text.strip()
    return " ".join(text.split())


# def remove_accented_chars(text):
#     # remove accented characters from text, e.g. cafÃ©
#     text = unidecode.unidecode(text)
#     return text


def expand_contractions(text):
    # expand shortened words, e.g. don't to do not
    text = contractions.fix(text)
    return text

def text_p(text):
    text = remove_whitespace(text)
#     text = remove_accented_chars(text)
    text = expand_contractions(text)
    text = text.lower()

    # print(text)
    return text
```

```python
#Removing whitespaces, accented characters and expanding contractions
df['PREPROCESSED_TEXT'] = df['PREPROCESSED_TEXT'].apply(text_p)
df
```

```python
# Punctuations removal
df['PREPROCESSED_TEXT'] = df['PREPROCESSED_TEXT'].apply(lambda x: re.sub(r'[^\w\s]', ' ', x + " "))
df
```

```python
# for i in range(len(df['PREPROCESSED_TEXT'])):
#   df['PREPROCESSED_TEXT'][i] = text_p(df['PREPROCESSED_TEXT'][i])
```

```python
# Tokenisation
df['PREPROCESSED_TEXT'] = df['PREPROCESSED_TEXT'].apply(word_tokenize)
df
```

```python
# Spelling correction
def spelling_correction(word):
    word = Word(word)
    result = word.correct()
    return result

for sentence in tqdm(df['PREPROCESSED_TEXT']):
    for i in range(len(sentence)):
        sentence[i] = spelling_correction(sentence[i])
df

```

```python
#Converting back to sentence to store in new csv file.

df["P_SENTENCE"] = df["PREPROCESSED_TEXT"]

for i in range(len(df["PREPROCESSED_TEXT"])):
    ProcessedSentence = ""
    for w in range(len(df["PREPROCESSED_TEXT"][i])):
        ProcessedSentence += df["PREPROCESSED_TEXT"][i][w]+" "
    df["P_SENTENCE"][i] = ProcessedSentence
df
```

```python
bigram_list = []
def bigram(processed_tokens):
    for i in range(len(processed_tokens)):
        if(len(processed_tokens[i])>=2):
            for j in range(1, len(processed_tokens[i])):
                temp_list = []
                temp_list.append(processed_tokens[i][j-1])
                temp_list.append(processed_tokens[i][j])
                bigram_list.append(temp_list)
```

```python
df_normalized = pd.DataFrame(np.ones((len(vocab), len(vocab))), columns=vocab, index=vocab)
for col in tqdm(df2):
    total=unigram[col]+len(vocab)
    for row in df2:
        score = sia.polarity_scores(col+" "+row)
        beta = score['neg']+score['pos']
        df2[col][row] = (df2[col][row]/total)+beta
        df_normalized[col][row] = df2[col][row]/total
df2
```

```python
def perplexity(sentence):
    l= sentence.split()
    pp=1
    for w in range(1, len(l)):
        pp *= 1/df2[l[w-1]][l[w]]
    perplexity_score=pow(pp, 1/len(l))
    return perplexity_score
    
```

```python
gen_df["PERPLEXITY"] = gen_df["TEXT"].apply(perplexity)
gen_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score

def train_and_evaluate(train_sentences, train_labels, test_sentences, test_labels):
    '''
    parameters:
    train_sentences : list of training sentences
    train_labels : list of training labels
    test_sentences : list of test sentences
    test_labels : list of test labels
    output:
    accuracy : accuracy of the test set
    '''
    # Model building
    model = make_pipeline(TfidfVectorizer(), MultinomialNB())
    # Training the model with the training data
    model.fit(train_sentences, train_labels)
    # Predicting the test data categories
    predicted_test_labels = model.predict(test_sentences)
    return accuracy_score(test_labels, predicted_test_labels)
```

```python
def preprocessing(sentence):
    sentence = re.sub(r'@[\w]*', '', sentence)
    sentence = re.sub(r'https?:\/\/\S+', '', sentence)
    sentence = re.sub(r"www\.[a-z]?\.?(com)+|[a-z]+\.(com)", '', sentence)
    sentence = text_p(sentence)
    sentence = re.sub(r'[^\w\s]', ' ', sentence + " ")
    sentence = word_tokenize(sentence)
    for i in range(len(sentence)):
        sentence[i] = spelling_correction(sentence[i])
    ProcessedSentence = ""
    for w in range(len(sentence)):
        ProcessedSentence += sentence[w]+" "
    sentence = ProcessedSentence
    return sentence
```
